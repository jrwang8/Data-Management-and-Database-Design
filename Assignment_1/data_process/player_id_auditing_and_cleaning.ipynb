```python
import pandas as pd
import numpy as np
```

Read the information of players

```python
data_player_id = pd.read_csv('player_id.csv')
data_player_id.head(10)
```

Cut the first columns

```python
data_player_id = data_player_id.drop(data_player_id.columns[0], axis=1)
```

Check the missing of dataset

```python
data_player_id.apply(lambda x: sum(x.isnull()),axis=0)
```

Find that there are 4 missing values in the FIRST_NAME, let's find it

```python
np.where(pd.isnull(data_player_id['FIRST_NAME']))
```

```python
data_player_id.ix[np.where(pd.isnull(data_player_id['FIRST_NAME']))[0],:]
```

We remember Sun. Yue was also a NBA player who is from China, so we think that his data might also have mistake.

```python
data_player_id[data_player_id['LAST_NAME'] == 'Sun']
```

Correct these mistakes

```python
data_player_id.ix[1706,'FIRST_NAME'] = 'Nene'

data_player_id.ix[3159,'FIRST_NAME'] = 'Qi'
data_player_id.ix[3159,'LAST_NAME'] = 'Zhou'

data_player_id.ix[4335,'FIRST_NAME'] = 'Ming'
data_player_id.ix[4335,'LAST_NAME'] = 'Yao'

data_player_id.ix[4341,'FIRST_NAME'] = 'Jianlian'
data_player_id.ix[4341,'LAST_NAME'] = 'Yi'

data_player_id.ix[3751, 'FIRST_NAME'] = 'Yue'
data_player_id.ix[3751, 'LAST_NAME'] = 'Sun'

data_player_id.ix[[1706,3159,4335,4341,3751]]
```

Recheck the missing data

```python
data_player_id.apply(lambda x: sum(x.isnull()),axis=0)
```

Perfect.
