Find that there are 20 missing data in the ORGANIZATION and ORGANIZATION_TYPE

```python
miss_org = np.where(pd.isnull(data_player_draft_history['ORGANIZATION']))
miss_org
```

```python
miss_org_tpye = np.where(pd.isnull(data_player_draft_history['ORGANIZATION_TYPE']))
miss_org_tpye
```

```python
data_player_draft_history.loc[miss_org[0],['ORGANIZATION', 'ORGANIZATION_TYPE']]
```

Change the NaN into 'Unknown'

```python
for i in miss_org[0]:
    data_player_draft_history.loc[i, 'ORGANIZATION'] = 'Unknown'
    data_player_draft_history.loc[i, 'ORGANIZATION_TYPE'] = 'Unknown'
data_player_draft_history.loc[miss_org[0],['ORGANIZATION', 'ORGANIZATION_TYPE']]
```

Perfect.

Change the order of the dataset and assign new draft id

```python
data_player_draft_history = data_player_draft_history.sort_values(by = ['SEASON','ROUND_NUMBER','ROUND_PICK'] , ascending = True)
data_player_draft_history = data_player_draft_history.reset_index(drop=True)

for i in range (0,data_player_draft_history.shape[0]): 
    data_player_draft_history.iat[i,0] = i+1
    
print(data_player_draft_history.head(10))
```
